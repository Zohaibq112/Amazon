```python
students_grades = [
    ('John', 85),
    ('Alice', 92),
    ('Bob', 78),
    ('Charlie', 95),
    ('David', 88)
]

students_courses = {
    'John': ['Math', 'Physics'],
    'Alice': ['Math', 'Computer Science'],
    'Bob': ['Physics', 'Chemistry'],
    'Charlie': ['Math', 'Biology'],
    'David': ['Computer Science', 'Chemistry']
}

grades = {name: grade for name, grade in students_grades}
course_data = {}

for student, courses in students_courses.items():
    for course in courses:
        if course not in course_data:
            course_data[course] = []
        course_data[course].append((student, grades[student]))

print("Average grades per course:")
for course, students in course_data.items():
    avg = sum(grade for _, grade in students) / len(students)
    
    Hgrade=max(students)
    print(f"{course}: {avg:}, highest gradev are :{Hgrade}")


    
```
